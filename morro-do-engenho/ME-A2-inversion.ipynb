# 3D inversion of the magnetic anomalies ME and A2 using fatiando.gravmag.harvester

Import required modues:

```python
%pylab inline
import json
from fatiando import gravmag, logger, mesher
from fatiando.vis import mpl, myv
from IPython.display import Image
```

```python
log = logger.get()
log.info(logger.header())
```

# The data

```python
x, y, z, topo, tf = loadtxt('ME-A2.dat', unpack=True, usecols=[2, 3, 4, 5, 6])

mpl.figure(figsize=(8,4))
mpl.axis('scaled')
mpl.title('ME and A2 anomalies (nT)')
mpl.contourf(y, x, tf, (100, 100), 40, interp=True)
mpl.colorbar()
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.m2km()
```

Some data properties:

```python
area = [x.min(), x.max(), y.min(), y.max()]
print "area:", area
print "max topo:", topo.max()
print "min topo:", topo.min()
```

Load the physical properties from the JSON file:

```python
with open('physical_properties.json') as f:
    props = json.load(f)
print props
```

# Inversion 1: ME and A2 with same physical property

Create a prism mesh:

```python
bounds = (x.min(), x.max(), y.min(), y.max(), -topo.max(), 15000)
mesh = mesher.PrismMesh(bounds, (30, 90, 60))
mesh.carvetopo(x, y, topo)
```

Wrap the data in containers that *harvester* understands:

```python
data = [gravmag.harvester.TotalField(x, y, z, tf, inc=props['inducing']['inclination'], dec=props['inducing']['declination'])]
```

Create the seeds:

```python
propA2 = dict(props['total'])
propME = dict(props['total'])
locations = [[19000, 38000, 3000, propA2],
             [8000,  17000, 0, propME]]
seeds = gravmag.harvester.sow(locations, mesh)

mpl.figure(figsize=(8,4))
mpl.axis('scaled')
mpl.title('ME and A2 anomalies (nT) + seed locations')
mpl.contourf(y, x, tf, (100, 100), 40, interp=True)
mpl.colorbar()
for s in locations:
    mpl.plot(s[1], s[0], 'ok')
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.m2km()
```

Run the inversion:

```python
estimate, predicted = gravmag.harvester.harvest(data, seeds, mesh, 120, 0.0003)

mesh.addprop('magnetization', estimate['magnetization'])
body = mesher.vremove(0, 'magnetization', mesh)

bottom = max(c.z2 for c in body)
print "Bottom at: %g m" % (bottom)
top = min(c.z1 for c in body)
print "Top at: %g m" % (top)
print "Mesh top: %g m" % (bounds[-2])

residuals = tf - predicted[0]
anomaly_range = tf.max() - tf.min()
print 'Range of total field anomaly:', anomaly_range
stddev = residuals.std()
print 'Residuals standard deviation:', stddev
print 'Percent of residuals standard deviation:', 100*float(stddev)/anomaly_range
```

## Results

Plot the adjustment and a histogram of the residuals:

```python
mpl.figure()
shape = (60, 60)
mpl.title("Observed = color | Predicted = contour")
mpl.axis('scaled')
levels = mpl.contour(y, x, tf, shape, 10, interp=True, color='r', style='dashed')
#mpl.colorbar()
mpl.contour(y, x, predicted[0], shape, levels, interp=True)
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.m2km()

mpl.figure()
mpl.title('Residuals: mean=%g stddev=%g' % (residuals.mean(), residuals.std()))
mpl.hist(residuals, bins=50, color='gray')
mpl.xlabel('Residuals (nT)')
mpl.ylabel('Number of occurences')
#mpl.xlim(-1000, 1000)

mpl.figure(figsize=(8,4))
mpl.axis('scaled')
mpl.contourf(y, x, residuals, shape, 15, interp=True)
mpl.colorbar()
mpl.contour(y, x, tf, shape, 10, interp=True)
```

# Inversion 2: A2 with larger magnetization

The predicted data is smaller for A2 (positive residuals).
So increase magnetization.

```python
mesh2 = mesher.PrismMesh(bounds, (30, 90, 60))
mesh2.carvetopo(x, y, topo)
# Alter the magnetization to seed what range works
seeds[0].props['magnetization'] = 20

estimate2, predicted2 = gravmag.harvester.harvest(data, seeds, mesh2, 120, 0.0003)

mesh2.addprop('magnetization', estimate2['magnetization'])
body2 = mesher.vremove(0, 'magnetization', mesh2)

bottom = max(c.z2 for c in body2)
print "Bottom at: %g m" % (bottom)
top = min(c.z1 for c in body2)
print "Top at: %g m" % (top)
print "Mesh top: %g m" % (bounds[-2])

residuals = tf - predicted2[0]
anomaly_range = tf.max() - tf.min()
print 'Range of total field anomaly:', anomaly_range
stddev = residuals.std()
print 'Residuals standard deviation:', stddev
print 'Percent of residuals standard deviation:', 100*float(stddev)/anomaly_range
```
