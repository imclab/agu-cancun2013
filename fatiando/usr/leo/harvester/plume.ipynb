```python
"""                                                                             
GravMag: 3D inversion in spherical coordinates using harvester and tesseroids   
"""
from IPython.display import Image
from fatiando import gravmag, gridder, logger, mesher, utils
from fatiando.vis import mpl, myv
import cPickle as pickle
```

```python
log = logger.get()
```

```python
props = {'density':1000}
model = [mesher.Tesseroid(-59, -51, -24, -16, -150000, -2700000, props)]
```

```python
scene = myv.figure(zdown=False, size=(800,800))
myv.tesseroids(model, 'density')
myv.earth(opacity=0.4)
myv.core(opacity=0.7)
myv.continents()
scene.scene.camera.position = [34583434.886585221, -14231179.759453828, -20486242.130268116]
scene.scene.camera.focal_point = [4.7340651334352174e-09, 0.0, -3.3138455934046522e-08]
scene.scene.camera.view_angle = 15.36
scene.scene.camera.view_up = [0.23696103470968202, -0.56345342622187156, 0.79143521813737283]
scene.scene.camera.clipping_range = [21601872.363654409, 69227028.81077157]
scene.scene.camera.compute_view_plane_normal()
scene.scene.render()
myv.savefig('plume-model.png')
myv.show()
```

```python
Image(filename='plume-model.png')
```

```python
bounds = [-70, -40, -40, 0, -100000, -3000000]
shape = (100, 100)
lons, lats, heights = gridder.regular(bounds[:4], shape, z=250000)
gz = utils.contaminate(gravmag.tesseroid.gz(lons, lats, heights, model), 0.01, percent=True)
```

```python
mpl.figure(figsize=(10, 8))
bm = mpl.basemap(bounds[:4], 'ortho')
bm.drawcoastlines()
bm.drawmapboundary()
mpl.contourf(lons, lats, gz, shape, 30, basemap=bm)
cb = mpl.colorbar()
cb.set_label('mGal')
mpl.savefig('plume-gravity.png')
```

```python
mesh = mesher.TesseroidMesh(bounds, (15, 40, 40))
```

```python
seeds = gravmag.harvester.sow([[-55, -20, -500000, props]], mesh)
```

```python
scene = myv.figure(zdown=False, size=(800, 800))
myv.tesseroids(model, style='wireframe', linewidth=2)
myv.tesseroids([mesh[s.i] for s in seeds])
myv.earth(opacity=0.4)
myv.core(opacity=0.7)
myv.continents()
scene.scene.camera.position = [34583434.886585221, -14231179.759453828, -20486242.130268116]
scene.scene.camera.focal_point = [4.7340651334352174e-09, 0.0, -3.3138455934046522e-08]
scene.scene.camera.view_angle = 15.36
scene.scene.camera.view_up = [0.23696103470968202, -0.56345342622187156, 0.79143521813737283]
scene.scene.camera.clipping_range = [21601872.363654409, 69227028.81077157]
scene.scene.camera.compute_view_plane_normal()
scene.scene.render()
myv.savefig('plume-seed.png')
myv.show()
```

```python
Image(filename='plume-seed.png')
```

```python
data = [gravmag.harvester.Gz(lons, lats, heights, gz, meshtype='tesseroid')]
```

```python
estimate, predicted = gravmag.harvester.harvest(data, seeds, mesh, compactness=10, threshold=0.001)
```

```python
mesh.addprop('density', estimate['density'])
```

```python
result = mesher.vremove(0, 'density', mesh)
```

```python
top = max(t.top for t in result)
bottom = min(t.bottom for t in result)
print "top:", top
print "bottom:", bottom
```

```python
mpl.figure(figsize=(20, 8))
mpl.subplot(1, 2, 1)
bm = mpl.basemap(bounds[:4], 'ortho')
bm.drawcoastlines()
bm.drawmapboundary()
levels = mpl.contour(lons, lats, gz, shape, 8, basemap=bm, color='k')
mpl.contour(lons, lats, predicted[0], shape, levels, basemap=bm, color='r')
mpl.subplot(1, 2, 2)
mpl.title('residuals')
mpl.hist(gz - predicted[0], bins=20)
mpl.savefig('plume-fit.png')
```
