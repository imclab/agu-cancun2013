# Analysis of the sensitivity of the solution to changes in the seeds

```python
%pylab inline
import sys
sys.path.append('../fatiando')
import json
import cPickle as pickle
from fatiando import gravmag, gridder, mesher, utils
from fatiando.vis import mpl, myv
```

## Generate a model

```python
bounds = [-500, 500, -500, 500, 0, 1000]
props = {'magnetization':utils.ang2vec(10, -40, -90)}
props = {'magnetization':10}
model = [mesher.Prism(-200,   0, -100, 100,  200,  400, props),
         mesher.Prism(-100, 100, -100, 100,  400, 600, props),
         mesher.Prism(   0, 200, -100, 100,  600, 800, props)]
```

Now lets take a look at our model.

```python
def view(scene, name):
    scene.scene.camera.position = [1454.0205679139528, -2986.2532751475665, 113.6010127605271]
    scene.scene.camera.focal_point = [-6.7775605414220763, -12.940259236832674, 584.37798758388476]
    scene.scene.camera.view_angle = 24.0
    scene.scene.camera.view_up = [-0.04829731745280999, 0.13301219219893073, -0.98993693024015295]
    scene.scene.camera.clipping_range = [1850.7679575833986, 5228.8734103499155]
    scene.scene.camera.compute_view_plane_normal()
    scene.scene.render()
    myv.savefig('../presentation/fig/spine-%s.png' % (name))
```

```python
scene = myv.figure(size=(700, 700))
myv.prisms(model, 'magnetization')
myv.axes(myv.outline(bounds), ranges=[b*0.001 for b in bounds], nlabels=3, fmt='%.1f')
myv.wall_east(bounds)
myv.wall_bottom(bounds)
view(scene, 'model')
myv.show()
```

## Forward model the synthetic data

Now that we have our model, we can produce synthetic noise corrupted data by forward modeling.

```python
inc, dec = -45, -10
area = bounds[:4]
shape = (25, 25)
x, y, z = gridder.regular(area, shape, z=-150)
tf = utils.contaminate(gravmag.prism.tf(x, y, z, model, inc, dec), stddev=5) # stddev is error in nT

mpl.figure(figsize=(5, 4))
mpl.axis('scaled')
mpl.contourf(y, x, tf, shape, 30, cmap=mpl.cm.seismic)
cb = mpl.colorbar()
cb.set_label('nT')
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.m2km()
mpl.savefig('../presentation/fig/spine-data.png')
```

## Inversion with 1 seed at the top

```python
mesh = mesher.PrismMesh(bounds, (30, 30, 30))
print 'Mesh dimensions:', mesh.dims
print 'Mesh size:', mesh.size
#locations = [[ -100, 0, 210, props], [ 0, 0, 410, props], [ 100, 0, 610, props]]
#locations = [[ -100, 0, 210, props], [ 0, 0, 410, props]]
locations = [[ -100, 0, 210, props]]
seeds = gravmag.harvester.sow(locations, mesh)
```

```python
scene = myv.figure(size=(700, 700))
myv.prisms(model, 'magnetization', opacity=0.3)
myv.prisms([mesh[s.i] for s in seeds])
myv.axes(myv.outline(bounds), ranges=[b*0.001 for b in bounds], nlabels=3, fmt='%.1f')
myv.wall_east(bounds)
myv.wall_bottom(bounds)
view(scene, 'seed')
myv.show()
```

Run the inversion.

```python
data = [gravmag.harvester.TotalField(x, y, z, tf, inc, dec)]
estimate, predicted = gravmag.harvester.harvest(data, seeds, mesh, compactness=0.3, threshold=0.002) # single seed
#estimate, predicted = gravmag.harvester.harvest(data, seeds, mesh, compactness=0.5, threshold=0.001) # 2 seeds
#estimate, predicted = gravmag.harvester.harvest(data, seeds, mesh, compactness=1, threshold=0.0005) # 3 seeds
#estimate, predicted = gravmag.harvester.harvest(data, seeds, mesh, compactness=3.5, threshold=0.0005)
mesh.addprop('magnetization', estimate['magnetization'])
bodies = mesher.vremove(0, 'magnetization', mesh)
print "# of prisms in the solution:", len(bodies)
residuals = tf - predicted[0]
print "Residuals mean:", residuals.mean()
print "Residuals standard deviation:", residuals.std()
```

```python
mpl.figure(figsize=(5, 4))
mpl.title('Observed (color) vs predicted (black)')
mpl.axis('scaled')
levels = mpl.contourf(y, x, tf, shape, 10, cmap=mpl.cm.seismic)
cb = mpl.colorbar(orientation='vertical')
cb.set_label('nT')
mpl.contour(y, x, predicted[0], shape, levels, color='k')
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.m2km()
mpl.savefig('../presentation/fig/spine-oneseed-fit.png')
```

```python
scene = myv.figure(size=(700, 700))
plot = myv.prisms(model, 'magnetization', style='wireframe', linewidth=3)
plot.actor.mapper.scalar_visibility = False
myv.prisms([mesh[s.i] for s in seeds], 'magnetization')
myv.prisms(bodies, 'magnetization')
myv.axes(myv.outline(bounds), ranges=[b*0.001 for b in bounds], nlabels=3, fmt='%.1f')
myv.wall_east(bounds)
myv.wall_bottom(bounds)
view(scene, 'oneseed')
myv.show()
```

## Experiments

```python
def spine(mesh):
    nz, ny, nx = mesh.shape
    locations = []
    for layer in mesh.layers():
        nonzero = mesher.vremove(0, 'magnetization', layer)
        if len(nonzero) > 15:
            centerofmass = sum(p.center() for p in nonzero)/len(nonzero)
            locations.append(centerofmass.tolist() + [props])
    return locations
```

```python
locations = spine(mesh)
seeds = gravmag.harvester.sow(locations, mesh)
scene = myv.figure(size=(700, 700))
myv.prisms(model, 'magnetization', opacity=0.3)
myv.prisms([mesh[s.i] for s in seeds])
myv.axes(myv.outline(bounds), ranges=[b*0.001 for b in bounds], nlabels=3, fmt='%.1f')
myv.wall_east(bounds)
myv.wall_bottom(bounds)
view(scene, 'spine')
myv.show()
```

```python
estimate, predicted = gravmag.harvester.harvest(data, seeds, mesh, compactness=3, threshold=0.0005)
mesh.addprop('magnetization', estimate['magnetization'])
bodies = mesher.vremove(0, 'magnetization', mesh)
print "# of prisms in the solution:", len(bodies)
residuals = tf - predicted[0]
print "Residuals mean:", residuals.mean()
print "Residuals standard deviation:", residuals.std()
```

```python
mpl.figure(figsize=(5, 4))
mpl.title('Observed (color) vs predicted (black)')
mpl.axis('scaled')
levels = mpl.contourf(y, x, tf, shape, 10, cmap=mpl.cm.seismic)
cb = mpl.colorbar()
cb.set_label('nT')
mpl.contour(y, x, predicted[0], shape, levels, color='k')
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.m2km()
mpl.savefig('../presentation/fig/spine-final-fit.png')
```
