# Invert both Morro do Engenho (ME) and the A2 anomalies

```python
%pylab inline
import sys
sys.path.append('../fatiando')
import json
import cPickle as pickle
from fatiando import gravmag, mesher, utils, constants
from fatiando.vis import mpl, myv
```

## Load the observed data and the physical properties

```python
x, y, z, height, tf = loadtxt('../data/ME-A2.dat', unpack=True, usecols=[2, 3, 4, 5, 6])
shape = (60, 60)
area = [x.min(), x.max(), y.min(), y.max()]
with open('../data/physical_properties.json') as f:
    props = json.load(f)
inc, dec = props['inducing']['inclination'], props['inducing']['declination']
# Calculate the induced magnetization
induced = utils.ang2vec(utils.nt2si(props['inducing']['total field']), inc, dec)*props['susceptibility']/constants.PERM_FREE_SPACE
# and the remanent
remanent = utils.ang2vec(props['remanent']['intensity'], props['remanent']['inclination'], props['remanent']['declination'])
magnetization = induced + remanent
print "Magnetization [intensity (A/m), inc, dec]:", utils.vec2ang(magnetization)
print "Regional inclination, declination:", [inc, dec]
```

```python
mpl.figure(figsize=(5, 5))
mpl.axis('scaled')
mpl.title('Total field magnetic anomaly')
mpl.contourf(y, x, tf, shape, 30, interp=True, cmap=mpl.cm.seismic)
cb = mpl.colorbar(orientation='horizontal')
cb.set_label('nT')
mpl.m2km()
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.savefig('../presentation/fig/me-data.png', dpi=600)

mpl.figure(figsize=(5, 5))
mpl.axis('scaled')
mpl.title('Topography')
mpl.contourf(y, x, height, shape, 30, interp=True)
cb = mpl.colorbar(orientation='horizontal')
cb.set_label('meters')
mpl.m2km()
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
```

## Mesh and seeds

First we need to make a mesh and some seeds.

```python
bounds = area + [-height.max(), 15000]
mesh = mesher.PrismMesh(bounds, (30, 50, 35))
mesh.carvetopo(x, y, height)
print "Mesh dimensions (x, y, z):", mesh.dims
print "Mesh size:", mesh.size

locations = [[9000, 17000, 2000, {'magnetization':magnetization}],
             [20000, 38000, 3000, {'magnetization':magnetization}]]
seeds = gravmag.harvester.sow(locations, mesh)

mpl.figure(figsize=(5, 5))
mpl.axis('scaled')
mpl.title('Seed locations')
mpl.contourf(y, x, tf, shape, 30, interp=True, cmap=mpl.cm.seismic)
cb = mpl.colorbar(orientation='horizontal')
cb.set_label('nT')
for s in locations:
    mpl.plot(s[1], s[0], 'ok')                 
mpl.m2km()
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.savefig('../presentation/fig/me-seed-map.png', dpi=600)
```

## Inversion

```python
data = [gravmag.harvester.TotalField(x, y, z, tf, inc, dec)]
estimate, predicted = gravmag.harvester.harvest(data, seeds, mesh, compactness=20, threshold=0.0005)
mesh.addprop('magnetization', estimate['magnetization'])
bodies = mesher.vremove(0, 'magnetization', mesh)
residuals = tf - predicted[0]
print "Residuals mean:", residuals.mean()
print "Residuals standar deviation:", residuals.std()
```

Plot the fit, residuals, and estimate.

```python
mpl.figure(figsize=(5, 5))
mpl.axis('scaled')
mpl.title('Observed (color) vs Predicted (black)')
levels = mpl.contourf(y, x, tf, shape, 25, cmap=mpl.cm.seismic, interp=True)
cb = mpl.colorbar(orientation='horizontal')
cb.set_label('nT')
mpl.contour(y, x, predicted[0], shape, levels, color='k', interp=True)
mpl.m2km()
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.savefig('../presentation/fig/me-fit.png', dpi=600)

mpl.figure(figsize=(5, 5))
mpl.axis('scaled')
mpl.title('Zoom of ME anomaly')
levels = mpl.contourf(y, x, tf, shape, 25, cmap=mpl.cm.seismic, interp=True)
#cb = mpl.colorbar(orientation='vertical')
#cb.set_label('nT')
mpl.contour(y, x, predicted[0], shape, levels, color='k', interp=True)
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.xlim(9000, 25000)
mpl.ylim(0, 20000)
mpl.m2km()
mpl.savefig('../presentation/fig/me-fit-zoom-me.png', dpi=600)

mpl.figure(figsize=(5, 5))
mpl.axis('scaled')
mpl.title('Zoom of A2 anomaly')
levels = mpl.contourf(y, x, tf, shape, 25, cmap=mpl.cm.seismic, interp=True)
#cb = mpl.colorbar(orientation='vertical')
#cb.set_label('nT')
mpl.contour(y, x, predicted[0], shape, levels, color='k', interp=True)
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.xlim(30000, 45000)
mpl.ylim(15000, 27000)
mpl.m2km()
mpl.savefig('../presentation/fig/me-fit-zoom-a2.png', dpi=600)

mpl.figure(figsize=(5, 5))
mpl.axis('scaled')
mpl.title('Residuals')
mpl.pcolor(y, x, residuals, shape, vmin=-abs(residuals).max(), vmax=abs(residuals).max(), cmap=mpl.cm.seismic, interp=True)
cb = mpl.colorbar(orientation='horizontal')
cb.set_label('nT')
mpl.m2km()
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
```

## Use data weights to get rid of the "tentacles"

Try using data weights to remove the "tentacles" that try to fit the geologic noise in the data. 
The weights decrease as the data point gets farther away from the seeds.
If the weight is 1, then changing the estimate has a large influence on that data point.
If the weight is 0, then changing the estimate has low influence on that data point.

To do this, we need to create another data object with the parameters `influence` and `seeds` set.
`influence` sets how far spread is the influence of each body that will be grown from a seed.

```python
data = [gravmag.harvester.TotalField(x, y, z, tf, inc, dec, influence=5000, seeds=seeds)]
```

Lets have a look at the weight of each data point.

```python
mpl.figure(figsize=(5, 5))
mpl.axis('scaled')
mpl.title('Weights')
mpl.pcolor(y, x, data[0].weights, shape, interp=True)
cb = mpl.colorbar(orientation='horizontal')
mpl.m2km()
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.savefig('../presentation/fig/me-weights.png', dpi=600)
```

Now run the inversion.

```python
estimate, predicted_weights = gravmag.harvester.harvest(data, seeds, mesh, compactness=25, threshold=0.0005)
mesh.addprop('magnetization', estimate['magnetization'])
bodies_weights = mesher.vremove(0, 'magnetization', mesh)
residuals_weights = tf - predicted_weights[0]
print "Residuals mean:", residuals_weights.mean()
print "Residuals standar deviation:", residuals_weights.std()
```

```python
mpl.figure(figsize=(5, 5))
mpl.axis('scaled')
mpl.title('Observed (color) vs Predicted (black)')
levels = mpl.contourf(y, x, tf, shape, 25, cmap=mpl.cm.seismic, interp=True)
cb = mpl.colorbar(orientation='horizontal')
cb.set_label('nT')
mpl.contour(y, x, predicted_weights[0], shape, levels, color='k', interp=True)
mpl.m2km()
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.savefig('../presentation/fig/me-weight-fit.png', dpi=600)

mpl.figure(figsize=(5, 5))
mpl.axis('scaled')
mpl.title('Zoom of ME anomaly')
levels = mpl.contourf(y, x, tf, shape, 25, cmap=mpl.cm.seismic, interp=True)
#cb = mpl.colorbar(orientation='vertical')
#cb.set_label('nT')
mpl.contour(y, x, predicted_weights[0], shape, levels, color='k', interp=True)
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.xlim(9000, 25000)
mpl.ylim(0, 20000)
mpl.m2km()
mpl.savefig('../presentation/fig/me-weight-fit-zoom-me.png', dpi=600)

mpl.figure(figsize=(5, 5))
mpl.axis('scaled')
mpl.title('Zoom of A2 anomaly')
levels = mpl.contourf(y, x, tf, shape, 25, cmap=mpl.cm.seismic, interp=True)
#cb = mpl.colorbar(orientation='vertical')
#cb.set_label('nT')
mpl.contour(y, x, predicted_weights[0], shape, levels, color='k', interp=True)
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.xlim(30000, 45000)
mpl.ylim(15000, 27000)
mpl.m2km()
mpl.savefig('../presentation/fig/me-weight-fit-zoom-a2.png', dpi=600)
```
