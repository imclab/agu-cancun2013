# Perform the inversion of the magnetic anomaly ME using fatiando.gravmag.harvester

Setup:

```python
%pylab inline
import sys
sys.path.append('../fatiando')
import json
import cPickle as pickle
from fatiando import gravmag, logger, mesher, gridder
from fatiando.vis import mpl, myv
from IPython.display import Image
```

```python
log = logger.get()
log.info(logger.header())
```

## The data

```python
x, y, z, topo, tf = loadtxt('ME.dat', unpack=True, usecols=[2, 3, 4, 5, 6])
tf-=11.1166
mpl.figure(figsize=(8,4))
mpl.axis('scaled')
mpl.title('ME anomaly (nT)')
mpl.contourf(y, x, tf, (100, 100), 30, interp=True)
mpl.colorbar()
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.m2km()
```

Some data properties:

```python
area = [x.min(), x.max(), y.min(), y.max()]
print "area:", area
print "max topo:", topo.max()
print "min topo:", topo.min()
print "sum of the data:", tf.sum()
```

Load the physical properties from the JSON file:

```python
with open('physical_properties.json') as f:
    props = json.load(f)
print props
```

### The inversion

Create a prism mesh:

```python
bounds = (x.min(), x.max(), y.min(), y.max(), -topo.max(), 15000)
mesh = mesher.PrismMesh(bounds, (16, 35, 30))
mesh.carvetopo(x, y, topo)
print "dims [x, y, z]:", mesh.dims
print "size:", mesh.size
```

Wrap the data in containers that *harvester* understands:

```python
data = [gravmag.harvester.TotalField(x, y, z, tf, inc=props['inducing']['inclination'], dec=props['inducing']['declination'])]
```

Create the seeds:

```python
# Alter the magnetization to seed what range works
propME = dict(props['total'])
locations = [[8000,  17000, 1, propME]]
#locations = [[8000,  17000, 1, propME], [17000,  16000, 500, propME]]
seeds = gravmag.harvester.sow(locations, mesh)

mpl.figure(figsize=(5,3.3))
mpl.axis('scaled')
mpl.title('ME anomaly')
mpl.contourf(y, x, tf, (100, 100), 60, interp=True)
cb = mpl.colorbar()
cb.set_label('nT')
for s in locations:
    mpl.plot(s[1], s[0], 'ok')
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.m2km()
mpl.savefig('ME-data_seeds.png')
```

Run the inversion:

The value of the regularization parameter (compactness) was chosen as the smallest one that gave a compact estimate.

```python
estimate, predicted = gravmag.harvester.harvest(data, seeds, mesh, 700, 0.001)

mesh.addprop('magnetization', estimate['magnetization'])
body = mesher.vremove(0, 'magnetization', mesh)

bottom = max(c.z2 for c in body)
print "Bottom at: %g m" % (bottom)
top = min(c.z1 for c in body)
print "Top at: %g m" % (top)
print "Mesh top: %g m" % (bounds[-2])

residuals = tf - predicted[0]
anomaly_range = tf.max() - tf.min()
print 'Range of total field anomaly:', anomaly_range
stddev = residuals.std()
print 'Residuals standard deviation:', stddev
print 'Residuals mean:', residuals.mean()
print 'Percent of residuals standard deviation:', 100*float(stddev)/anomaly_range
```

Plot the adjustment and a histogram of the residuals:

```python
#mpl.figure(figsize=(20, 4))
shape = (100, 100)
mpl.figure(figsize=(5, 4))
#mpl.subplot(1, 3, 1)
#mpl.title("Observed = color | Predicted = contour")
mpl.axis('scaled')
levels = mpl.contour(y, x, tf, shape, 12, interp=True, color='r')
#mpl.colorbar()
mpl.contour(y, x, predicted[0], shape, levels, interp=True)
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.m2km()
mpl.savefig('ME-fit.png')

mpl.figure(figsize=(5, 4))
#mpl.subplot(1, 3, 3)
mpl.hist(residuals, bins=20, color='gray')
mpl.text(300, 500, r'$\mu = %g$' % (residuals.mean()), fontsize=15)
mpl.text(300, 450, r'$\sigma = %g$' % (residuals.std()), fontsize=15)
mpl.xlabel('Residuals (nT)')
mpl.ylabel('Number of')
#mpl.xlim(-1000, 1000)
mpl.savefig('ME-residuals-hist.png')

mpl.figure(figsize=(5, 4))
#ax = mpl.subplot(1, 3, 2)
mpl.title("Residuals")
mpl.axis('scaled')
mpl.contourf(y, x, residuals, shape, 60, interp=True)
cb = mpl.colorbar(shrink=0.8)
cb.set_label('nT')
#mpl.annotate('Interfering source?', xy=(16000, 17000), xytext=(1000, 13000), horizontalalignment='left', verticalalignment='top',
#    arrowprops={'fill':None, 'shrink':0.05}, fontsize=12)
mpl.text(1000, 24000, 'Interfering source?', fontsize=12)
gca().arrow(8000, 23000, 6000, -5000, head_width=1000, head_length=1000, fc='k', ec='k')
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.m2km()
mpl.savefig('ME-residuals.png')
```

There is a dipole in the residuals! Could be an interfering source

Plot a North-South cross-section to see if the dipole in the residuals really appears there:

```python
profile_at = 16000
sub = [0.0, 27683.33898, profile_at, 160000]
xp, yp, tfp = gridder.interp(x, y, tf, (1, 50), area=sub)
xp, yp, predp = gridder.interp(x, y, predicted[0], (1, 50), area=sub)
xp, yp, predp, tfp = xp[0], yp[0], predp[0], tfp[0]

mpl.figure(figsize=(5, 3))
mpl.subplots_adjust(right=0.99, left=0.16)
mpl.title('Cross-section at y = %g km' % (profile_at*0.001))
mpl.grid(True)
mpl.plot(xp*0.001, tfp, '-r', label='Observed', linewidth=2)
mpl.plot(xp*0.001, predp, '-k', label='Predicted', linewidth=2)
mpl.legend(fancybox=True, shadow=True, loc='lower right')
mpl.text(10.5, -800, 'Interfering source?', fontsize=12)
gca().arrow(15, -500, 1, 490, head_width=0.5, head_length=100, fc='k', ec='k')
mpl.xlabel('North (km)')
mpl.ylabel('Total field anomaly (nT)')
mpl.xlim(xp.min()*0.001, xp.max()*0.001)
mpl.savefig('ME-cross_section.png')
```
